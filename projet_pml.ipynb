```python
import pandas as pd
data = pd.read_csv("data/file.csv", encoding='latin-1')
data=data[data['Area']=='China']
object_columns = data.select_dtypes(include='object').columns
data_1 = pd.get_dummies(data= data, columns=object_columns)
```

```python
data_1
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
```

```python
# Define your training data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = data_1.drop("yield_amound", axis=1)
y = data_1['yield_amound']
scaler = StandardScaler().fit(X)
Xstd=scaler.transform(X)


X_train, X_test, y_train, y_test = train_test_split(Xstd, y, test_size=0.2, random_state=42)
```

```python
# Preprocess the data
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))  # Reshape input for RNN
X_train
```

```python
# Build the RNN model
model = Sequential()
model.add(SimpleRNN(units=125, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
from keras.layers import LSTM, Dense, Dropout, Masking, Embedding

model_lstm = Sequential()
model_lstm.add(LSTM(units=125, activation="tanh", input_shape=(X_train.shape[1],1)))
model_lstm.add(Dense(units=1))

model_lstm.compile(optimizer="RMSprop", loss="mse")
```

```python
# Train the model
model_lstm.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Train the mode
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
# Test the model

X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))  # Reshape input for RNN
y_pred = model.predict(X_test)

print("Predictions:", y_pred)

loss = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
```
